# 새 섹션

```python
from models.networks import define_G
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from models.networks import define_G
import torchvision.transforms as transforms
import torch

img_name = "img_main_akita_dog_sp.jpg"
img = Image.open(img_name)
plt.imshow(img), plt.show()

model = define_G(3, 3, 64, 'resnet_9blocks', 'instance', False, 'normal', 0.02, [])
model.load_state_dict(torch.load('300_net_G_A.pth'))
model.eval()

img = img.convert('RGB')

transformer = []
transformer.append(transforms.Resize((256,256),3))
transformer.append(transforms.ToTensor())
transformer.append(transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5)))
transformer = transforms.Compose(transformer)

img = transformer(img)
pred = model(img.unsqueeze(0))
pred = pred[0].detach().numpy().transpose(1,2,0)
plt.imshow(pred), plt.show()
```
